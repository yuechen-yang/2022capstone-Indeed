```python
import matplotlib.pyplot as plt
import seaborn as sns

from transformers import pipeline
import pandas as pd
```

```python
review_data_whole=pd.read_csv("Indeed_review.csv")
review_data_whole.head()
```

change the index of review data

```python
review_data=review_data_whole.iloc[1653595:2839014,:]
```

```python
review_data
```

run the model

```python
review_data["review_sentiment_label"]=""
review_data["review_sentiment_score"]=""
review_data.head()
```

sentiment_analysis

```python
classifier=pipeline("sentiment-analysis")
```

```python
review_data
```

Model

```python
i=0
```

```python
for item,row in review_data.iterrows():
  try:
    review_content=row["review_content"]
    #print(item)
    #print(i)
    #print(review_content)
    #print("*********************************")
    
    sen_content=classifier(review_content)
    sen_content=sen_content[0]
    review_data.iloc[i,9]=sen_content["label"]
    review_data.iloc[i,10]=sen_content["score"]
    # write to csv per 200 row
    if i >0 and i%200==0:
        print(i)
        print("*********************************")
        temp_write_review=review_data.iloc[i-200:i,:]
        temp_write_review.to_csv("sentiment_analysis_1653595.csv", mode='a', index=True, header=False, encoding='utf-8_sig')
    
    i=i+1
  except Exception as e:
    print(e)
    review_data.iloc[i,9]=""
    review_data.iloc[i,10]=""
    
    # write to csv per 10 row
    if i >0 and i%200==0:
        print(i)
        print("#####################3")
        temp_write_review=review_data.iloc[i-200:i,:]
        temp_write_review.to_csv("sentiment_analysis_1653595.csv", mode='a', index=True, header=False, encoding='utf-8_sig')
    
    i=i+1
```
