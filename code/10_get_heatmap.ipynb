```python
import pandas as pd
import matplotlib
import seaborn as sns
import random
```

```python
review_data=pd.read_csv("Indeed_review.csv")
company_detail_data=pd.read_csv("../Eikon/company_refinitiv_detail_data.csv")
```

```python
review_data
```

```python
company_detail_data
```

change to quarter

```python
review_data['review_date']= pd.to_datetime(review_data['review_date'])
```

```python
review_data['review_quarter'] = review_data['review_date'].dt.quarter
```

sample 10000 data

```python
#sub_sample_review=review_data.sample(n=100000)
sub_sample_review=review_data
```

```python
print(len(sub_sample_review))
sub_sample_review
```

Sort by Number of Employees

```python
company_name=sub_sample_review.merge(company_detail_data,how="left")
company_name=company_name[["company_name","Number of Employees"]]
company_name=company_name.drop_duplicates()
company_name
```

aggregate by quater

```python
temp=sub_sample_review[["company_name","review_quarter"]]
temp["flag"]=1
```

```python
agg_data=temp.groupby(["company_name","review_quarter"]).count()
agg_data
```

```python
agg_data=agg_data[agg_data["flag"]<1000]
agg_data=agg_data[agg_data["flag"]>100]
```

```python
agg_data=agg_data.reset_index()
```

```python
#agg_data=agg_data.sort_values(by='Number of Employees', ascending=False)
#agg_data=agg_data[["company_name","review_quarter","flag"]]
agg_data
```

change dataframe structure

```python
agg_data=agg_data.pivot(index='company_name', columns='review_quarter', values='flag')
agg_data
```

```python
company_name=company_name.rename(columns={"company_name":"new_name"})
company_name
```

```python
agg_data["new_name"]=agg_data.index
agg_data=agg_data.merge(company_name,how="left")
```

```python
agg_data=agg_data[agg_data["Number of Employees"].notnull()]
```

```python
agg_data=agg_data.sort_values(by='Number of Employees', ascending=False)
agg_data=agg_data[["new_name",1,2,3,4]]
agg_data
```

```python
agg_data=agg_data.set_index("new_name")
agg_data
```

```python
agg_data=agg_data.fillna(0)
agg_data
```

```python
sns.heatmap(agg_data)
```
